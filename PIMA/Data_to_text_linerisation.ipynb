```python
s=''
s1=''
l=[]
l1=[]
def data_to_text(maxValues,temp) :
    arr=[]
    string=''
    for index in range (temp.shape[0]):
        string+='<table>'
        for col_num in range(2):
            
            string+='<cell>'+str(maxValues.iloc[index][col_num])+'<col-header>'+temp.iloc[index][col_num]+'</col-header>'+'</cell>'
        string+='</table>'
        arr.append(string)
        string=''
    return arr     
        
l=data_to_text(maxValuesDiabetic,temp)
l1=data_to_text(maxValuesHealthy,temp1)
```

```python
np.column_stack((l, df[0]['Explanation'].values)).shape
```

```python
text_data_Diabetic=pd.DataFrame(np.column_stack((l, df[0]['Explanation'].values)),columns=['Input','Output'])
text_data_Healthy=pd.DataFrame(np.column_stack((l1, df[1]['Explanation'].values)),columns=['Input','Output'])
```
